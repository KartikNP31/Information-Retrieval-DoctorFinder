# This is the main code for crawling

```python
with open("./PractoData.csv", "a", newline="", encoding="utf-8") as csvfile:
    fieldnames = [
        "Doctor Name",
        "Experience",
        "Speciality",
        "Fees",
        "Rating",
        "Locality",
        "City",
    ]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for city in newCities:
        # for speciality in new_specialities:
        #     formatted_speciality = (
        #         speciality.replace(" ", "-")
        #         .replace("/", "-")
        #         .replace("(", "")
        #         .replace(")", "")
        #     )

        page = 1
        has_content = True
        while has_content:
            # print(f"Scraping page {page} for {speciality} in {city} city")
            current_url = (
                url.replace("${city}", city)
                # .replace("${speciality}", formatted_speciality)
                # .replace("${page}", str(page))
            )
            if check_url(current_url):
                response = requests.get(current_url)
                if response.status_code == 200:
                    print(current_url)
                    htmlContent = response.content
                    soup = BeautifulSoup(htmlContent, "html.parser")

                    list_name = soup.find_all("h2", class_="doctor-name")

                    if len(list_name) > 0:
                        doctor_name = []
                        doctor_experience = []
                        doctor_fees = []
                        doctor_city = []
                        doctor_speciality = []
                        doctor_rating = []
                        doctor_locality = []
                        doctor_description = []

                        grey_3_divs = soup.find_all("div", class_="u-grey_3-text")
                        for t in grey_3_divs:
                            div_span_elements = t.find("div", class_="u-d-flex")
                            if (
                                div_span_elements
                            ):  # Check if div_span_elements is not None
                                span_element = div_span_elements.find("span")
                                if span_element:  # Check if span_element is not None
                                    span_text = span_element.get_text()
                                    doctor_speciality.append(span_text)
                        list_fees = soup.find_all("div", class_="uv2-spacer--xs-top")
                        list_rating = soup.find_all(
                            "span", class_="o-label--success u-bold"
                        )
                        list_locality = soup.find_all(
                            "div", class_="u-bold u-d-inlineblock u-valign--middle"
                        )

                        # for t in list_speciality:
                        #     doctor_speciality.append(t.get_text())

                        for t in list_name:
                            doctor_name.append(t.get_text())

                        fees_consultation = []
                        for t in list_fees:
                            if re.search(r"\d+", t.get_text()):
                                fees_consultation.append(
                                    re.search(r"\d+", t.get_text()).group()
                                )
                        for t in fees_consultation:
                            if int(t) < 60:
                                doctor_experience.append(t)
                            else:
                                doctor_fees.append(t)

                        doctor_city.extend([city] * len(list_name))
                        # doctor_speciality.extend([speciality] * len(list_name))

                        for t in list_rating:
                            doctor_rating.append(t.get_text())

                        for t in list_locality:
                            doctor_locality.append((t.get_text()).split(",")[0])

                        length_of_features = [
                            len(doctor_city),
                            len(doctor_experience),
                            len(doctor_fees),
                            len(doctor_locality),
                            len(doctor_name),
                            len(doctor_speciality),
                            len(doctor_rating),
                        ]
                        lowest_length = min(length_of_features)
                        doctors = []

                        for i in range(lowest_length):
                            doctors.append(
                                {
                                    "Doctor Name": doctor_name[i],
                                    "Experience": doctor_experience[i],
                                    "Speciality": doctor_speciality[i],
                                    "Fees": doctor_fees[i],
                                    "Rating": doctor_rating[i],
                                    "Locality": doctor_locality[i],
                                    "City": doctor_city[i],
                                }
                            )

                        writer.writerows(doctors)
                        page += 1
                    else:
                        has_content = False  # No content found, exit the loop
                else:
                    print(
                        f"Failed to fetch page {page} for {speciality} in {city}. Status code: {response.status_code}"
                    )
                    has_content = False  # Break the loop if request fails
            else:
                print(f"Invalid URL: {current_url}")
                has_content = (
                    False  # Break the loop if URL is invalid or an error occurs
                )
```
